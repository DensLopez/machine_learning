# **9. Árboles de decisión**

En esta clase vamos a hablar sobre clasificadores **Random Forest**.

## **Componentes**

**Decisión tree [Arboles de decisión]**



Imaginemos un juguete para nuestro sobrino y queremos evitar que tenga bordes afilados, por lo que lo definiremos como "Save" que significa que no tiene bordes afilados.

Así que decimos que un juguete es **Bad (B)** si tiene bordes afilados y **Good (G)** si no los tiene. Entonces cada juguete obtendrá una puntuación **B o G**, dependiendo de la característica de los bordes.

En la primera fila de juguetes podemos ver diferentes formas del juguete. Por lo que un **Árbol de decisión** haría una serie de preguntas acerca de nuestro dataset, estas basadas en las **features** que tenemos.

Por ejemplo una de las **features** puede ser el **color**. Así que tal vez pregunte **¿Es azul?** y así mis datos se separan en 2 colores diferentes.
1. El grupo que no es azul, es decir el que pertenece al color rojo.
2. El grupo que es azul.

Después podemos realizar otra pregunta ¿Es círculo?, así nuevamente se vuelven a separar mis datos.

Un **nodo de decisión [decision node]** corresponde específicamente a la pregunta que será respondida y un **nodo hoja [leave node]** corresponde al grupo resultante que genera el **nodo de decisión** que se respondió previamente.

### **¿Cómo podemos hacer predicciones?**

Vayamos a otro ejemplo



Puedo realizar un conjunto de preguntas de manera similar. Por ejemplo; para el circulo rosa puedo preguntar ¿Es azul?



Entonces al contestar la pregunta lo clasifica.

Pero imaginemos que al realizar otra pregunta asociada con el azul. Se queda la respuesta como **Bad**. Y esto a pesar de ser un circulo lo clasifica como malo, porque en mi **árbol de decisión** que planteamos al inicio, el ser **azul** se considera como un **Bad (B)**.

Entonces podemos preguntar.



Aquí es donde recurrimos a **Random Forest**

Un **Random Forest** es un arreglo, grupo o conjunto de **Arboles de decisión** que trabajan para encontrar una respuesta adecuada o correcta.

Así que tener un arreglo de varios **Arboles de decision** podemos preguntar efectivamente de manera diferente acerca de nuestro dataset.

Para cada **data point** del conjunto se hace un proceso de **votación** en los que los arreglos de **arboles de decision** van a puntuar por la respuesta correcta.



### **Parámetros**

* **Número de árboles:** Nos dice el número de modelos de decision que tenemos y que son capaces para votar. Mientras más árboles tienes más robusta es la respuesta que obtienes, y hay menos varianza. Esto es debido a que los arboles no están correlacionados; es decir que uno no depende del otro. **Sin embargo;** mientras más árboles tenemos, más recursos de computo se necesitan para poder ejecutar estos modelos.

* **Máximo numero de features:** Es el número posible de elementos que queremos tener para poder ser capaz de realizar la predicción que queremos hacer.

* **Máx depth:** Este es el número de preguntas que el modelo va  a hacer (El número de niveles de un árbol de decisión). Mientras más preguntas tengamos, más profundo (depth) es el árbol.



* **$N_dplit$:** Esto es básicamente cuantos datos necesitamos en un **nodo de decisión** para ser capaz de realizar las preguntas.

* **$n_min$:** Es el número mínimo de data points que tenemos antes de terminar una pregunta.



### **Función de coste y regla de actualización**

+ **Función de coste:** La meta de la función es encontrar efectivamente la mejor forma de hacer **split** en el dataset, se apoya del criterio de entropía para poder encontrar el **threshold**.

+ **Regla de actualización:** Lo que nos dice de manera efectiva es, si nosotros tenemos **n** puntos de datos.



### **Evaluación**

Se pueden usar métrica de **clasificación** o **regresión** para evaluar que tan bueno es el modelo. Es decir como las vimos en anteriores modelos. **Regresión lineal** y **regresión logística**

**Matriz de confusión** para clasificación para encontrar valores predichos, es decir nuestras etiquetas en el dataset. Por otro lado.

**$\frac{MSE}{R^2}$** Esto es para tratar de predecir valores numéricos.



### **Repaso**

1. **Proceso de decisión:** En este paso lo que realizamos es escoger un feature y realizar un split de los datos, para separar los datos.
2.  **Función de coste:** Esta se va a encargar de preguntar ¿cuál es el **threshold** capaz de separar los datos de un **feature** establecido.
3.  **Regla de actualización:** Si no tenemos un valor mínimo de datos ($n_min$ values) continua haciendo preguntas, lo que significa; escoge otra **feature** para poder separar los datos.
