# **7. Regresin lineal**

Es una forma de obtener valores numricos a partir de **features** de nuestro dataset.

La regresin lineal tiene como objetivo dibujar una linea que describa el comportamiento de un **output (salida) o target** que est contenida dentro de **features** del dataset.



**Tipos de relaciones**
1. Relacin positiva: Ambas variables tienen un comportamiento de crecimiento.
2. Relacin negativa: Tienen un comportamiento opuesto, mientras una variable crece; la otra decrece.
3. Sin embargo la regresin lineal no es una buena eleccin para datos con comportamiento complejo, como se puede observar en la grfica **no lineal**.

### **Proceso de decisin**
Cuales son los parmetros que se necesitan y que significan en la regresin lineal.

El objetivo de los parmetros es observar cuales son los **pesos(weighs)** o **features** en los valores de entrada y el **bias($W_o$)**. Entonces la meta es encontrar estos valores, para poder ser capaz de predecir el valor.

Podemos pensar sobre estos **weights** como **hyphothesis** es decir; que tan fuerte es la relacin entre **input features** y **output**

As como en **$W_o$** que es el **intercepto** nos responde a la pregunta: **qu pasa si en nuestro feature hay un valor de 0 o no existe? qu valor de resultado esperaremos?**

### **Regla de actualizacin**
Cmo hacer que nuestro modelo obtenga mejores parmetros.



Esta regla trata de encontrar cuales valores ($w_0,w_1$) o en general **weighs (pesos)** son capaces en conjunto de minimizar la distancia de los **puntos reales** y **la prediccin**. 

La regla numrica usar una optimizacin numrica, para ser capaz de encontrar la mejor solucin de **weighs** posible para crear la linea de inters (regresin lineal).

### **Rendimiento**
Una vez que tenemos los 3 parmetros anteriores queremos preguntar cuando mi modelo de ML est trabajando bien?.

En el case de regresin lineal podemos usar **Mean Square Error (MSE)** que es un valor de la **funcin de error** o ocupar el $R^2$ que es la correlacin entre el valor de entrada y el valor de salida que predecimos.

Con estos valores podemos decir si existe una correlacin fuerte o una correlacin dbil.

Por ejemplo valores cercanos a:
- **1:** Buena correlacin positiva
- **0:** Correlacin dbil
- **-1:** Buena correlacin negativa
