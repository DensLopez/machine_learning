# **7. Regresión lineal**

Es una forma de obtener valores numéricos a partir de **features** de nuestro dataset.

La regresión lineal tiene como objetivo dibujar una linea que describa el comportamiento de un **output (salida) o target** que está contenida dentro de **features** del dataset.



**Tipos de relaciones**
1. Relación positiva: Ambas variables tienen un comportamiento de crecimiento.
2. Relación negativa: Tienen un comportamiento opuesto, mientras una variable crece; la otra decrece.
3. Sin embargo la regresión lineal no es una buena elección para datos con comportamiento complejo, como se puede observar en la gráfica **no lineal**.

### **Proceso de decisión**
Cuales son los parámetros que se necesitan y que significan en la regresión lineal.

El objetivo de los parámetros es observar cuales son los **pesos(weighs)** o **features** en los valores de entrada y el **bias($W_o$)**. Entonces la meta es encontrar estos valores, para poder ser capaz de predecir el valor.

Podemos pensar sobre estos **weights** como **hyphothesis** es decir; que tan fuerte es la relación entre **input features** y **output**

### **Regla de actualización**
Cómo hacer que nuestro modelo obtenga mejores parámetros.



Esta regla trata de encontrar cuales valores ($w_0,w_1$) o en general **weighs (pesos)** son capaces en conjunto de minimizar la distancia de los **puntos reales** y **la predicción**. 

La regla numérica usará una optimización numérica, para ser capaz de encontrar la mejor solución de **weighs** posible para crear la linea de interés (regresión lineal).

### **Rendimiento**
Una vez que tenemos los 3 parámetros anteriores queremos preguntar ¿cuando mi modelo de ML está trabajando bien?.

En el case de regresión lineal podemos usar **Mean Square Error (MSE)** que es un valor de la **función de error** o ocupar el $R^2$ que es la correlación entre el valor de entrada y el valor de salida que predecimos.

Con estos valores podemos decir si existe una correlación fuerte o una correlación débil.

Por ejemplo valores cercanos a:
- **1:** Buena correlación positiva
- **0:** Correlación débil
- **-1:** Buena correlación negativa



## Quick Review
**El proceso de decisión** es la combinación lineal de **weighs** e **inputs** sumando a el **bias**

Este es el valor que nuestro modelo va a predecir para un **valor numérico** dadas ciertas **features**



**Función de error** nos responde que tan bueno fue el acercamiento de nuestra predicción al valor real. Idealmente sería el mismo valor, por lo que **la función de error** tiene que ser muy pequeño.



**Regla de actualización** así como calculamos este valor, queremos encontrar el mejor ajuste de **weighs (W_n)** que juntos son capaces de predecir el valor más cercano a nuestro **valor de salida e interés (target)**.
