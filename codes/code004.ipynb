# 4. Tipos de modelos de Machine Learning

En esta clase hablaremos sobre diferentes modelos de ML y el tipo de tareas o problemas que resuelven.

La meta para esta clase es saber que tipo de aproximacin debo realizar y que tipo de problemas se van a resolver.

En la amplia categora de algoritmos de Machine Learning nos enfocaremos en 2:
- **Supervisados** 
- **No supervisados**

### **Aprendizaje supervisado**
La meta de este tipo de aprendizaje es realizar una ingesta de datos y tomar **features - variables** para predecir un **target - variable**. Esto puede ser muchas o pocas variables.
- **Regresin lineal**: El objetivo es predecir un **target** numrico o valor. Esto significa una variable cuantitativa. 
- **Clasificacin**: Queremos ser capaces de predecir una **etiqueta - label o grouping**. El objetivo es predecir una variable categrica.

Veamos algunos ejemplos de aplicacin que podemos construir para aplicar esto.



#### **Regresin**  
- Pronsticos de temperatura: Tal vez seamos capaces de predecir la temperatura que hay afuera,**features** nos ayudaran a predecir el valor numrico; es decir que tan clido o fri esta el clima.
- Prediccin de la popularidad de un producto: Podemos predecir la popularidad de un producto por el volumen o cantidad de ventas, ya sea analizando y segmentando demograficamente. Inclusive podemos predecir el comportamiento de la popularidad; incrementara, se mantendr o decrementar. Dependiendo de la estrategia que tomemos.
- Estimacin de la esperanza de vida
- Pronsticos de mercado

#### **Clasificacin**
- Retencin de clientes:
- Predecir fraude: Analizando si es una buena transaccin o no, basado en la ubicacin, tiempo y hasta posiblemente en patrones desconocidos.

### **Aprendizaje no supervisado**
Por el otro lado si el objetivo es desconocido. Lo que queremos es encontrar estructura y grupos dentro de los datos.
- **Clustering**: agrupar datos similares.
- **Reduccin de dimensionalidad**: Cuando queremos encontrar la menor cantidad posible de **features - variables** que son las ms importantes y capaces de caracterizar el comportamiento de los datos.

Esto nos dice cuales **features** son las ms correlacionadas entre si y cuales son diferentes entre si, cuando estamos tratando de encontrar grupos en nuestros datos.



#### **Clustering**
- Segmentacin de clientes: A que tipo de consumidores deberamos dirigir cierto anuncio, basado en su comportamiento, el sitio web, el tipo de productos que ve.
- Sistemas de recomendacin: Para saber que tipo de pelculas son recomendadas a una persona basado en su historial de vistas y tambin en los gustos comunes de la gente.

#### **Reduccin de dimensionalidad**
El objetivo es encontrar los **features** ms importantes que representan los grupos en los datos.
- Visualizacin de big data: Tu podras visualizar muchas columnas y filas, algunas veces millones. Y entonces tu podras crear una versin comprimida de estos **features**, que sera el equivalente de los grupos que hay en los datos.
