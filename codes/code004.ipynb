# 4. Tipos de modelos de Machine Learning

En esta clase hablaremos sobre diferentes modelos de ML y el tipo de tareas o problemas que resuelven.

La meta para esta clase es saber que tipo de aproximación debo realizar y que tipo de problemas se van a resolver.

En la amplia categoría de algoritmos de Machine Learning nos enfocaremos en 2:
- **Supervisados** 
- **No supervisados**

### **Aprendizaje supervisado**
La meta de este tipo de aprendizaje es realizar una ingesta de datos y tomar **features - variables** para predecir un **target - variable**. Esto puede ser muchas o pocas variables.
- **Regresión lineal**: El objetivo es predecir un **target** numérico o valor. Esto significa una variable cuantitativa. 
- **Clasificación**: Queremos ser capaces de predecir una **etiqueta - label o grouping**. El objetivo es predecir una variable categórica.

Veamos algunos ejemplos de aplicación que podemos construir para aplicar esto.



#### **Regresión**  
- Pronósticos de temperatura: Tal vez seamos capaces de predecir la temperatura que hay afuera,**features** nos ayudaran a predecir el valor numérico; es decir que tan cálido o frió esta el clima.
- Predicción de la popularidad de un producto: Podemos predecir la popularidad de un producto por el volumen o cantidad de ventas, ya sea analizando y segmentando demograficamente. Inclusive podemos predecir el comportamiento de la popularidad; incrementara, se mantendrá o decrementará. Dependiendo de la estrategia que tomemos.
- Estimación de la esperanza de vida
- Pronósticos de mercado

#### **Clasificación**
- Retención de clientes:
- Predecir fraude: Analizando si es una buena transacción o no, basado en la ubicación, tiempo y hasta posiblemente en patrones desconocidos.

### **Aprendizaje no supervisado**
Por el otro lado si el objetivo es desconocido. Lo que queremos es encontrar estructura y grupos dentro de los datos.
- **Clustering**: agrupar datos similares.
- **Reducción de dimensionalidad**: Cuando queremos encontrar la menor cantidad posible de **features - variables** que son las más importantes y capaces de caracterizar el comportamiento de los datos.

Esto nos dice cuales **features** son las más correlacionadas entre si y cuales son diferentes entre si, cuando estamos tratando de encontrar grupos en nuestros datos.



#### **Clustering**
- Segmentación de clientes: A que tipo de consumidores deberíamos dirigir cierto anuncio, basado en su comportamiento, el sitio web, el tipo de productos que ve.
- Sistemas de recomendación: Para saber que tipo de películas son recomendadas a una persona basado en su historial de vistas y también en los gustos comunes de la gente.

#### **Reducción de dimensionalidad**
El objetivo es encontrar los **features** más importantes que representan los grupos en los datos.
- Visualización de big data: Tu podrías visualizar muchas columnas y filas, algunas veces millones. Y entonces tu podrías crear una versión comprimida de estos **features**, que sería el equivalente de los grupos que hay en los datos.


## **Algoritmos de ML**



#### Supervisado
- Regresión: 
  - Regresión lineal
- Clasificación:
  - Regresión logística
  - Random forest

#### No supervisado
- Clustering: 
  - K- means
  - Hierarchical clustering
- Reducción de dimensionalidad:
  - T-SNE
  - Principal Component Analisys - PCA

## Resumen



### **NOTA**
Cuando no tengamos idea de un target, probablemente queramos encontrar patrones, por lo tanto el **Aprendizaje no supervisado** es el más apropiado.
