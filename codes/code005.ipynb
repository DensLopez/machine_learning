### **3a: Representando patrones en tus datos - histogramas**

Primero revisa las distribuciones de frecuencia. Para esto utilizars un **histograma**.

Un histograma usa **bins** o lmites de un valor numrico. Contamos cuntas filas caen dentro de los lmites superior e inferior de cada bin y esto crea la distribucin.

Pandas puede automatizar la creacin de histogramas para no contar desde cero. Puedes darle el nmero de bins o esquinas de los bins (el criterio para lmites inferior y superior).

Para nuestro caso utiliza el predeterminado (que generalmente es 10 bins).

Dado que queremos aprender el estado de si la persona tiene o no diabetes, podemos preguntar si nuestros datos estn bien distribuidos en valores de 1 o 0 para el `outcome`. Necesitamos clases igualmente balanceadas para hacer predicciones decentes.

Revisa esto con la grfica de barras resultado del cdigo siguiente:

````python
#Ajustar el tamao de fuente
plt.rcParams['font.size'] = 15

#Crear una figura y ajusta su tamao
f = plt.figure(figsize=(8,4))

#Crear un subplot o subtrama - al ser una sola figura 
#es 1,1,1.
ax = f.add_subplot(1,1,1)

#Grfica tus datos usando `hist`. Pasa el objeto 'ax' a Pandas. 
#Agrega un borde negro con un groso de 2.
data["Outcome"].hist(ax=ax, edgecolor='black', linewidth=2)

#Establece los lmites en el eje x
ax.set_xlim([-0.5, 1.5])

#Establece la frecuencia de tick. 
# Tenemos 0 y 1 que corresponden a S y No respectivamente.
ax.set_xticks([0, 1])

#Etiquetar xtick labels. 
ax.set_xticklabels(['No', 'S'])

#Crea el ttulo
ax.set_title('Diabetes? [Si - No]')

#Establece la etiqueta del eje x
ax.set_xlabel('Respuesta')

#Establece la etiqueta del eje Y
ax.set_ylabel('Frecuencia')

#Establece los lmites superior / inferior del eje Y
ax.set_ylim([0, 510])

#Hace que el resultado sea mejor visualmente.
#No es necesario, pero se ajusta al tamao de la figura
f.tight_layout
````

De lo anterior, podemos ver que tenemos ms "No" que "S", lo que significa que hay ms personas sin diabetes en este conjunto de muestra que con diabetes.

Intentemos esto con una columna ms complicada, como la edad. Nos gustara ver el rango de edad de los pacientes con los que estamos trabajando. Esto puede deberse a que los pacientes mayores y los ms jvenes tienen necesidades mdicas diferentes.

Por ejemplo, podemos usar un histograma para observar especficamente la edad de nuestros pacientes:

````python
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, edgecolor='black', linewidth=1)
ax.set_title("Age range of patients")
ax.set_ylim([0, 410])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
````

Si resolviste lo anterior, la mayora de nuestros pacientes parecen tener entre 20 y 40 aos. Algunos son mayores. Es posible que deseemos probar diferentes valores de bin.

### **3b: Representando patrones en tus datos - grficas de dispersin (scatterplots)**

Los histogramas muestran datos en conjunto de una manera conveniente basada en conteos. A veces, es posible que quieras ver cmo determinadas caractersticas (features) o columnas tienen relaciones entre s.

Para hacer esto usamos un diagrama o grfica de dispersin. 

Una grfica de dispersin muestra cada punto de datos de su conjunto de datos como una relacin entre dos features que especificas.

Recuerda que esto se hace tomando 2 columnas del dataset y graficando, para cada fila, el valor de $(f_1, f_2)$ donde $f$ representa un feature o columna en los ejes ***x*** y ***y*** respectivamente.

Crearemos una grfica entre ``BloodPressure (x)`` y ``Age (y)``. Para esto asumimos que las personas jvenes tiene presiones de sangre (BloodPressure) ms bajas.

````python
plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["Age"], data["BloodPressure"], alpha=0.25)
ax.set_title("Blood pressure vs. age range of patients")
ax.set_ylabel("Blood pressure")
ax.set_xlabel("Age")
f.tight_layout()
````

Nota algo interesante aqu! Hay varias personas con 0 presin arterial! Desde el punto de vista mdico esto es muy extrao.

Los diagramas de dispersin nos ayudan a comprender cmo nuestros datos pueden tener informacin inapropiada o faltante. En el caso anterior, es poco probable que las personas tengan 0 presin arterial, pero es ms probable que el profesional mdico no tuviera estos datos listos o no se midieron. No queremos entrenar modelos con este tipo de informacin.

Para lo siguientes podemos considerar estos puntos como **outliers** o **anomalas**. Para este caso sera apropiado removerlas del dataset para asegurar que entrenamos nuestro modelo con solo los valores reportados completos.

Para realizar lo anterior usaremos Pandas para solo [seleccionar](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#boolean-indexing) las filas de nuestro dataset donde BloodPressure > 0.

````python
# Seleccionar las filas donde el valor es mayor a 0.
data = data[data["BloodPressure"] > 0]
print("Patients with blood pressure\n", data.shape) # Rows x Columns
````

Removimos 35 registros que tenan esta anomala. Volvamos a grficar y observar los datos.

````python
plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["Age"], data["BloodPressure"], alpha=0.25)
ax.set_title("Blood Pressure v. Age range of patients")
ax.set_ylabel("Blood Pressure")
ax.set_xlabel("Age")
f.tight_layout() # Makes things pretty
````

Ahora podemos ver que tenemos una tendencia un poco ms clara:

- Pacientes mayores tienden a tener ms probabilidades de tener una presin arterial ms alta.
- Pacientes ms jvenes pueden estar ms dispersos.
