### Definición 
Machine Learning es la ciencia de usar algoritmos para **identificar patrones en datos** con el fin de resolver un problema de interés. 

Ahora para ilustrar o clarificar este concepto, discutiremos más adelante 2 tipos de aprendizaje:

Pero la meta principal de los algoritmos de ML es recibir features o descriptores (variables) de un conjunto de datos y encontrar patrones.



En esta imagen $X_1$ y $X_2$ representan **features** arbitrarias de un dataset de interés. Las relaciones que podemos ver son:
- **Aprendizaje no supervisado**: Grupos
- **Aprendizaje supervisado**: Un patrón que parece sugerir cierta relación entre el eje $X_1$ y eje $X_2$.

De esa forma Machine Learning se especializa en ser capaz de encontrar que son estos patrones y cómo están ampliamente correlacionados. 



Los algoritmos de Machine Learning están en nuestro día a día, cómo por ejemplo en algoritmos que eviten emails de tipo fraudulentos. Se invierten una buena cantidad de dinero para evitar que las empresas se inunden con mensajes maliciosos. 

En 2015 Google consiguió identificar el 99.99% de emails tipo spam.

Veamos cómo logro hacerlo.



En la imagen anterior tenemos 2 textos:

1. Spam: En este podemos ver ciertos enlaces que nos redireccionan a un sitio web (este tipo de patrón se considera sospechoso).
2. Normal: Vemos un mensaje de agradecimiento.

Ahora, el mensaje es algo similar pero hay ciertas cosas que se consideran sospechosas o inapropiadas, por ejemplo; adjuntar un enlace con la finalidad de visitarlo y que tal vez conduzca a un sitio web fraudulento.

Los modelos de ML son capaces de detectar esto, recoger o encontrar estos patrones, para poder encontrar cual correo podría ser SPAM vs cual podría ser NORMAL.
