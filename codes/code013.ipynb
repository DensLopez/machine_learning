# **13. Cómo mejorar las predicciones de Redes Neuronales**

En esta clase vamos a hablar acerca de como mejorar **Neural Networks** específicamente entender como hacer nuestras predicciones robustas o fuertes y asegurar que estamos evaluando preguntas apropiadamente para poder hacer **Good predictions**. Así que quiero hablar de cómo los modelos hacen el ajuste **(fit)** de los datos.

## **Dropout**
Hay 3 maneras como son ajustados los datos.



Por supuesto la figura de en medio es lo que idealmente queremos, es decir lo que es el patrón realmente. Pero hay 2 casos limites que no son los mejores.

**Underfit:** Esto significa que el modelo realmente no está aprendiendo el patrón que hay en **training data** para ser capaz de realizar una buena predicción.

**Overfit:** Esto significa que el modelo esta memorizando los datos que lo que significa que en realidad no está aprendiendo patrones, pero está creando una forma de predecir sin realmente entender nada.

Una forma de proteger esto es realizar **Dropout**. Dropout es usado en el procedimiento de **entrenamiento** como una **capa especifica**, y lo que efectivamente hace es ignorar algunos nodos mientras se realizar el proceso de **entrenamiento**.

Tu puedes pensar que esto es **bloquear información** así que el modelo no tendrá suficiente información para ser capaz de memorizar, con esto podemos efectivamente preguntara **¿cómo creamos un modelo esperando que no haga overfit?**

### **Prevenir el overfit**



Por otro lado podemos usar **Performance (Rendimiento)** para evaluar si el modelo bueno.

Ahora cuando vayamos a traves del **proceso de entrenamiento** iremos realizando un proceso de calculo forward pass y backward pass, para cada **data point**.

Tan pronto como lo hagas para cada **data point** esto se considera una **época (epoch)**, y puedes hacerlo para multiples épocas hasta que el modelo aprenda el patrón y sea capaz de realizar una predicción. Sin embargo desde que las **Redes Neuronales** tienden a hacer **overfit** tiendes a ver que el **learning rate performance** incrementa a lo largo del tiempo.

Puedes usar un **validation dataset (dataset de validación)** para preguntar efectivamente ¿estos patrones son los que la Red Neuronal aprendió dentro del dataset de entrenamiento?.

Así que el **Optimal number of epoch (el valor óptimo de épocas)** que deberías usar es aquel en que la gráfica de **validación de rendimiento (performance)** peaks (tiene un pico) y no cambia más, por lo tanto tu puedes usar el modelo para evaluar tu dataset.
