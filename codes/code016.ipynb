## **16. ¿Qué mas aprender de Inteligencia Artificial?**

En esta clase vamos a hacer un resumen de nuestro curso y hablar un poco acerca **¿Cómo puedes obtener otras skills para ampliar tu entendimiento de machine learning?**

### **¿Qué aprendimos?**

Aprendimos mucho empezando con:
- Bases de Data Science, cómo cargar y trabajar la data. Todas las formas de visualización y relaciones de **features** en nuestro dataset.
- Algoritmos de Aprendizaje Supervisado: Para ser capaces de predecir **targets**, ambos en el contexto de **Regresión** y **Clasificación**.
- Algoritmos de Aprendizaje No Supervisado: Para identificar **patrones** o **estructuras** en nuestra data en particular set de **features**.
- Redes Neuronales: Potentes modeles capaces de realizar predicciones de funciones complejas, aprendimos a entrenarlas y evaluar su **performance**.
