# **Regresión logística**
En esta clase hablaremos sobre la **regresión logística** que es usada para problemas de clasificación.

Para explicar por qué, hagamos un recuento sobre la regresión lineal.



En la figura izquierda: **regresión lineal** es capaz de predecir valores numéricos por el ajuste de una linea en los puntos del dataset.

De manera similar **regresión logística** también esta ajustando una linea, pero en lugar de ajustarla a los puntos, **se trata de encontrar una linea para separar 2 clases diferentes**.

Así en la figura del lado derecho podemos ver; puntos azules y puntos rojos. Así que el objetivo de esta linea es **encontrar la linea que separa los puntos azules de los rojos**, con esto en mente vayamos al proceso de que hace la **regresión logística** para ser capaz de hacer esta predicción.



## **Ejemplo**


Supongamos que eres un maestro y estas tratando de recomendar cuantas horas deben estudiar los alumnos para poder pasar los exámenes. En la gráfica se trata de encontrar esa relación entre horas de estudio vs cuantos pasan.

### **Parámetros**


El objetivo de la regresión logística es encontrar parámetros, específicamente que sean capaces de predecir la probabilidad de pasar.

Así que en los ejes tenemos:
- **X**: Horas de estudio
- **Y**: Probabilidad de pasar

Así que típicamente podemos decir que un valor por encima de 0.5 es para decir que **aprobará** y un valor por debajo equivale a lo contrario.

Con esto en mente vamos a tratar de optimizar la predicción de la probabilidad.

Así que ahondemos en la **función de error o función de coste**

### **Función de coste**


El objetivo de la **función de coste** es poder evaluar si nuestra probabilidad predicha es correcta para aquellos que pasan, no pasan o fallan adecuadamente."

Los componentes de la **función de coste**  corresponden a predecir correctamente la probabilidad de aprobar (Likelihood of passing)contra predecir correctamente la probabilidad de fallar.

En este caso es una aproximación binaria, pero hay formas de adaptar esto a un **caso multiclase** así que también la **regresión logística** puede hacer esto.



Por último hay una forma de optimizar la probabilidad en la cual minimizamos, manipulando los parámetros para ser capaz de encontrar el mejor ajuste de valores así que la **regla de actualización** busca minimizar la **función de coste**; específicamente encontrando los valores de **weighs ($W_n$)** que mejor ajusten a la linea que separa los datos es comúnmente la opción para la **regla de actualización**.

### **Rendimiento**
Una vez dicho eso, ¿Cómo podemos saber que los modelos de **regresión logística** tienen un desempeño adecuado.



Una forma de hallarlo es mediante **Confusion matrix (Matriz de confusión)**. Y el objetivo de la matriz de confusión es poder evaluar si la predicción de los modelos representan realmente los valores de los datos.

Así que cuando estamos entrenando nuestro dataset de validación podemos específicamente preguntar sobre el número de valores positivos y valores negativos en un problema de clasificación de 2 clases.

Cuantos valores se predijeron de manera apropiada y cuando no fue así ¿cuál fué el formato de falla más probable? Fueron negativos, o positivos o viceversa.

El objetivo aquí es excepcionalmente bueno cuando tienes conjuntos de datos desbalanceados, lo que significa que tienes, por ejemplo, más estudiantes que aprueban que los que fallan, y por lo tanto podría reflejar que predijiste que todos aprueban en lugar de fallar.

**Pero** si tienes un dataset balanceado la aproximación más común capaz de describir el **rendimiento** de mejor manera es **Accuracy**; que es el conteo de cuantos veces se hizo una predicción de manera correcta (True Postive, True Negative) sobre todos lo datos del dataset.

**Idealmente, una alta precisión cercana a uno es algo por lo que deberías esforzarte**

## **Quick Review**


### **Proceso de decisión:** 
Aquí se está tratando de predecir la linea que nos ayudará a encontrar la probabilidad de que un estudiante pase. En este caso es **la probailidad de predecir una cierta clase en una forma muy general**.

### **La función de error o función de coste**
Se usa de manera para tratar de encontrar ciertos **weighs** que son capaces de predecir si alguien pasa o no de manera apropiada.

### **La regla de actualización**
Queremos encontrar valores **weight** que son los más apropiamos para encontrar la linea que divide o aquellos que optimicen la probabilidad.
