# **Regresin logstica**
En esta clase hablaremos sobre la **regresin logstica** que es usada para problemas de clasificacin.

Para explicar por qu, hagamos un recuento sobre la regresin lineal.



En la figura izquierda: **regresin lineal** es capaz de predecir valores numricos por el ajuste de una linea en los puntos del dataset.

De manera similar **regresin logstica** tambin esta ajustando una linea, pero en lugar de ajustarla a los puntos, **se trata de encontrar una linea para separar 2 clases diferentes**.

As en la figura del lado derecho podemos ver; puntos azules y puntos rojos. As que el objetivo de esta linea es **encontrar la linea que separa los puntos azules de los rojos**, con esto en mente vayamos al proceso de que hace la **regresin logstica** para ser capaz de hacer esta prediccin.



## **Ejemplo**


Supongamos que eres un maestro y estas tratando de recomendar cuantas horas deben estudiar los alumnos para poder pasar los exmenes. En la grfica se trata de encontrar esa relacin entre horas de estudio vs cuantos pasan.

### **Parmetros**


El objetivo de la regresin logstica es encontrar parmetros, especficamente que sean capaces de predecir la probabilidad de pasar.

As que en los ejes tenemos:
- **X**: Horas de estudio
- **Y**: Probabilidad de pasar

As que tpicamente podemos decir que un valor por encima de 0.5 es para decir que **aprobar** y un valor por debajo equivale a lo contrario.

Con esto en mente vamos a tratar de optimizar la prediccin de la probabilidad.

As que ahondemos en la **funcin de error o funcin de coste**

### **Funcin de coste**


El objetivo de la **funcin de coste** es poder evaluar si nuestra probabilidad predicha es correcta para aquellos que pasan, no pasan o fallan adecuadamente."

Los componentes de la **funcin de coste**  corresponden a predecir correctamente la probabilidad de aprobar (Likelihood of passing)contra predecir correctamente la probabilidad de fallar.

En este caso es una aproximacin binaria, pero hay formas de adaptar esto a un **caso multiclase** as que tambin la **regresin logstica** puede hacer esto.



Por ltimo hay una forma de optimizar la probabilidad en la cual minimizamos, manipulando los parmetros para ser capaz de encontrar el mejor ajuste de valores as que la **regla de actualizacin** busca minimizar la **funcin de coste**; especficamente encontrando los valores de **weighs ($W_n$)** que mejor ajusten a la linea que separa los datos es comnmente la opcin para la **regla de actualizacin**.

### **Rendimiento**
Una vez dicho eso, Cmo podemos saber que los modelos de **regresin logstica** tienen un desempeo adecuado.



Una forma de hallarlo es mediante **Confusion matrix (Matriz de confusin)**. Y el objetivo de la matriz de confusin es poder evaluar si la prediccin de los modelos representan realmente los valores de los datos.

As que cuando estamos entrenando nuestro dataset de validacin podemos especficamente preguntar sobre el nmero de valores positivos y valores negativos en un problema de clasificacin de 2 clases.

Cuantos valores se predijeron de manera apropiada y cuando no fue as cul fu el formato de falla ms probable? Fueron negativos, o positivos o viceversa.

El objetivo aqu es excepcionalmente bueno cuando tienes conjuntos de datos desbalanceados, lo que significa que tienes, por ejemplo, ms estudiantes que aprueban que los que fallan, y por lo tanto podra reflejar que predijiste que todos aprueban en lugar de fallar.

**Pero** si tienes un dataset balanceado la aproximacin ms comn capaz de describir el **rendimiento** de mejor manera es **Accuracy**; que es el conteo de cuantos veces se hizo una prediccin de manera correcta (True Postive, True Negative) sobre todos lo datos del dataset.

**Idealmente, una alta precisin cercana a uno es algo por lo que deberas esforzarte**
