# **3. Data Science framework**

Antes de poder entrenar modelos de ML, es importante ser capaz de entender e inspeccionar nuestros datos. Nuestros modelos serán tan buenos como lo sean nuestros datos, así que en esta clase nos enfocaremos en Ciencia de Datos (Data Science), específicamente la terminología fundamental, también repasaremos el manejo de data frame de datos para entenderlos y visualizarlos de manera gráfica.

Para empezar enfoquemos en conceptos clave:
- **Data - Datos**: Unidades de información o "hechos" de observaciones. Estos pueden ser desde cuestionarios o encuestas simples hasta información financiera (complejo) recolectada por bancos.
- **Features - variables**: Son descriptores sobre los tipos de información que tienen tus observaciones. Por ejemplo en un set de datos sobre parámetros medicos puede contener: estatura, peso, genero, nivel de azúcar, condition de salud, etc.

Ahora en nuestros datos existe un cierto tipo de formato que son filas y columnas.
- **Rows - Filas**: Representan cada observación o registro individual (muestras) de nuestros datos.
- **Columns - Columnas**: Representan cada variable o features de nuestros datos.

Los set de datos pueden ser pequeños o enormes, hasta miles de observaciones. Dependiendo de con que estés trabajando

- **Outlier**: En el contexto de Ciencia de Datos un outlier puede tener multiples significados. Puede ser algor extraño como punto(s) de datos o data point(s) que se comporta de forma extraña. Pero también puede ser un valor faltante o incorrecto. La meta es ser capaz de detectar que tipo de outlier es el que tenemos en nuestros datos y tratar de manera adecuada para poder darle un buen uso.
- **Pre-processing**: Prepara datos para su uso en un modelo de ML.Esto incluye detectar valores faltantes, eliminarlos o imputarlos para encontrar que valor podrían tener. Esto con la finalidad de entregar los datos a los modelos de machine learning y que me puedan proveer información útil.

### Tipos de datos

Ahora para analizar los datos, veamos los diferentes tipos de datos que hay. Los más populares son los **numéricos** y los **categóricos**

- Numéricos: Es un tipo de dato que es un numero y puede ser un entero o un decimal(valores continuos).
- Categóricos: Son features que son etiquetados; por ejemplo ponemos tener un One Hot encoding en el que es capaz de representar etiquetas con valores numéricos; algún ejemplo puede ser: **¿qué tipo de forma tiene un juguete?** *circulo* o *cuadrado*.

Datos más complejos incluyen **texto e imágenes**. estos tipos de datos requieren **pre-processing** más complejo. Hay que tener en cuenta que los modelos de ML pueden trabajar con este tipo de datos.

- **NaN (Not a Number)**: Esto es a menudo un valor faltante o un valor corrupto e incluso puede ser algo extraño en el data set. Es responsabilidad tuya que te descubras si es un valor faltante o algo extraño en el data set.
